```python
import wfdb
import os
```

```python
n_subjects = 1
n_sessions = 1
n_samples = 38

for i in range(n_subjects):
    for j in range(n_sessions):
        for k in range(n_samples):
            try:
                wfdb.dl_database('hd-semg',os.getcwd(), records=['pr_dataset/subject{:02}'.format(i+4) + '_session' + str(j+2) + '/maintenance_raw_sample'+str(k+38)])
            except:
                pass
```
